```python
import torch
import random
from diffusers import StableDiffusionPipeline
from PIL import Image
import numpy as np
import os
from google.colab import files


pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5", torch_dtype=torch.float16)
pipe = pipe.to("cuda")


def generate_one_image(prompt, pipe,seed=None):
    if seed is not None:
        torch.manual_seed(seed)
        np.random.seed(seed)
        random.seed(seed)
    image = pipe(prompt).images[0]
    return image

prompts = "creat a character call Bob who is a muscle boy 16 years old | Bob eat cake on the grass | Bob play basketball  "
images = {}

for prompt in prompts.split('|'):
    prompt = prompt.strip()  # 去除前后的空格
    images[prompt] = generate_one_image(prompt, pipe,10)

# 查看和保存图像
for prompt, image in images.items():
    # 显示图像
    image.show()

    # 保存图像到文件
    file_name = prompt.replace(" ", "_") + ".png"  # 生成文件名
    image.save(file_name)

# 确保文件保存后可以在Colab文件浏览器中看到
import os
os.listdir('.')
```
